```python
from pathlib import Path
import numpy as np
import sys
sys.path.append('../../analyze')
sys.path.append('../../lib_processing')
import create_heatmap
import utils
```

```python
book='alma'
chapter = '036'
```

```python
root_embeddings = Path('../../data/embeddings').absolute().resolve()
root_text = Path('../../data/text').absolute().resolve()
ne1 = root_embeddings / 'w040_i020' / 'scriptures' / 'book_of_mormon' / book
embeds = np.genfromtxt(ne1 / (chapter + '.embeddings.csv'), delimiter=',')
offs = np.genfromtxt(ne1 / (chapter + '.offsets.csv'), delimiter=',', dtype=int)
embed_means = np.genfromtxt(root_embeddings / '..' / 'means' / 'w040_i020' / 'scriptures' / 'mean.csv', delimiter='\n')
embeds_norm = embeds - embed_means
text = (root_text / 'scriptures' / 'book_of_mormon' / book / (chapter + '.txt')).read_text()
```

```python
print(root_text)
```

```python
cos_sim_norm = utils.cos_sim(embeds_norm, embeds_norm)
```

```python
create_heatmap.create_heatmap(cos_sim_norm, offs, offs, text, text, 0.2)
```

```python
cos_sim_cluster_max = create_heatmap.get_top_matches(cos_sim_norm, offs, offs, text, text, 0.4)
```

```python
create_heatmap.create_heatmap(cos_sim_cluster_max, offs, offs, text, text, 0.4)
```

```python
book_1 = 'book_of_mormon'
sub_book_1='alma'
chapter_1 = '032'

book_2 = 'new_testament'
sub_book_2='heb'
chapter_2 = '011'
```

```python
root_embeddings = Path('../../data/embeddings').absolute().resolve()
root_text = Path('../../data/text').absolute().resolve()
embed_means = np.genfromtxt(root_embeddings / '..' / 'means' / 'w040_i020' / 'scriptures' / 'mean.csv', delimiter='\n')

root_1 = root_embeddings / 'w040_i020' / 'scriptures' / book_1 / sub_book_1
embeds_1 = np.genfromtxt(root_1 / (chapter_1 + '.embeddings.csv'), delimiter=',')
offs_1 = np.genfromtxt(root_1 / (chapter_1 + '.offsets.csv'), delimiter=',', dtype=int)
embeds_1_norm = embeds_1 - embed_means
text_1 = (root_text / 'scriptures' / book_1 / sub_book_1 / (chapter_1 + '.txt')).read_text()

root_2 = root_embeddings / 'w040_i020' / 'scriptures' / book_2 / sub_book_2
embeds_2 = np.genfromtxt(root_2 / (chapter_2 + '.embeddings.csv'), delimiter=',')
offs_2 = np.genfromtxt(root_2 / (chapter_2 + '.offsets.csv'), delimiter=',', dtype=int)
embeds_2_norm = embeds_2 - embed_means
text_2 = (root_text / 'scriptures' / book_2 / sub_book_2 / (chapter_2 + '.txt')).read_text()
```

```python
cos_sim_norm = utils.cos_sim(embeds_1_norm, embeds_2_norm)
```

```python
create_heatmap.create_heatmap(cos_sim_norm, offs_1, offs_2, text_1, text_2, 0)
```

```python
cos_sim_cluster_max = create_heatmap.get_top_matches(cos_sim_norm, offs_1, offs_2, text_1, text_2, 0.35)
```

```python
create_heatmap.create_heatmap(cos_sim_cluster_max, offs_1, offs_2, text_1, text_2, 0.35)
```
